# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def travelling_salesman(cost_matrix):

    total_cities = len(cost_matrix)

    memo = {}

    def solve(current_city, visited_cities):

        if visited_cities == (1 << total_cities) - 1:
            return cost_matrix[current_city][0]

        if (current_city, visited_cities) in memo:
            return memo[(current_city, visited_cities)]

        minimum_cost = 999999

        for city in range(total_cities):

            if visited_cities & (1 << city) == 0:

                travel_cost = cost_matrix[current_city][city]

                next_visit = solve(
                    city,
                    visited_cities | (1 << city)
                )

                total_cost = travel_cost + next_visit

                if total_cost < minimum_cost:
                    minimum_cost = total_cost

        memo[(current_city, visited_cities)] = minimum_cost

        return minimum_cost

    answer = solve(0, 1)

    return answer
```

```python
cities = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

result = travelling_salesman(cities)

print("Lowest Cost is:", result)
```

# Exercise
1. Print path of cities visited
2. Take graph input from user
3. Test with different number of cities

```python
def travelling_salesman(graph):

    n = len(graph)

    memory = {}

    def solve(current_city, visited):

        if visited == (1 << n) - 1:
            return graph[current_city][0], [current_city, 0]

        if (current_city, visited) in memory:
            return memory[(current_city, visited)]

        minimum_cost = 999999
        best_path = []

        for next_city in range(n):

            if visited & (1 << next_city) == 0:

                cost, path = solve(
                    next_city,
                    visited | (1 << next_city)
                )

                total_cost = graph[current_city][next_city] + cost

                if total_cost < minimum_cost:
                    minimum_cost = total_cost
                    best_path = [current_city] + path

        memory[(current_city, visited)] = (minimum_cost, best_path)

        return memory[(current_city, visited)]

    return solve(0, 1)


cities = int(input("Enter number of cities: "))

graph = []

print("Enter the cost matrix row by row:")

for i in range(cities):

    row = list(map(int, input().split()))

    graph.append(row)

minimum_cost, path_taken = travelling_salesman(graph)

print("Minimum Travelling Cost:", minimum_cost)

print("Path Taken:")

for city in path_taken:
    print(city, end=" ")
```
