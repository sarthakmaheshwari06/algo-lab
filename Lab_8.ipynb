# Minimum Cost Spanning Tree
Problem Description:
A spanning tree of a graph can be defined as a graph with minimal set of edges that connect all vertices. A minimum spanning tree of a graph is a spanning tree of the graph with least weight (where the weight is computed by adding the weights of all the edges in the spanning tree). In general, a graph can have multiple minimum spanning trees. The problem is to find a minimum spanning tree of a graph.

This implementation uses a dictionary to represent the weighted graph, where the vertices are keys, and the values are dictionaries containing their adjacent vertices and their weights.

Please note that this implementation assumes that the input graph is connected and weighted.

First, let's create a simple graph to demonstrate Prim's algorithm:

```python
graph = {
    "A": {"B": 10, "C": 15, "D": 20},
    "B": {"A": 10, "C": 12},
    "C": {"A": 15, "B": 12, "D": 18},
    "D": {"A": 20, "C": 18},
}
```

Now, let's implement Prim's algorithm:

```python
def prim_algorithm(graph):
    # Initialize data structures
    min_tree_edges = []
    min_weight = 0
    visited = set()
    num_vertices = len(graph)

    if not graph:
        return min_tree_edges, min_weight

    # Start with an arbitrary vertex (e.g., the first one in the graph keys)
    start_vertex = list(graph.keys())[0]
    visited.add(start_vertex)

    while len(visited) < num_vertices:
        min_edge_weight = float('inf')
        best_u = None
        best_v = None

        # Iterate over all vertices currently in the MST
        for u in visited:
            # Iterate over all neighbors of 'u'
            for v, weight in graph[u].items():
                # If 'v' is not yet in the MST, and this edge is the smallest so far
                if v not in visited and weight < min_edge_weight:
                    min_edge_weight = weight
                    best_u = u
                    best_v = v

        if best_v is not None:
            visited.add(best_v)
            min_tree_edges.append((best_u, best_v, min_edge_weight))
            min_weight += min_edge_weight
        else:
            # If no new vertex can be added, it means the graph is not connected
            # (assuming the problem statement guarantees a connected graph).
            break

    return min_tree_edges, min_weight
```

Now, let's run the Prim's algorithm on the example graph:
Keep in mind that this example uses a simple representation of a graph, but for more complex cases, you might want to use a library such as NetworkX to represent and manipulate graphs.

```python
mst_edges_prim, min_weight_prim = prim_algorithm(graph)
print("Prim's Minimum Spanning Tree Edges: ", mst_edges_prim)
print("Prim's Minimum Weight: ", min_weight_prim)
```

## Kruskal's Algorithm
Kruskal's algorithm is another popular algorithm for finding the Minimum Spanning Tree (MST) of a connected, undirected graph. It's a greedy algorithm that works as follows:
1.  **Sort all edges** in non-decreasing order of their weights.
2.  **Initialize an empty MST** and a disjoint-set data structure (Union-Find) where each vertex is in its own set.
3.  **Iterate through the sorted edges**: For each edge `(u, v)` with weight `w`:
    *   If `u` and `v` are not already in the same connected component (checked using `find` operation), add the edge to the MST and merge the components of `u` and `v` (using `union` operation).
4.  **Stop** when `V-1` edges have been added to the MST, where `V` is the number of vertices.

This implementation uses a Union-Find data structure to efficiently detect cycles and merge components.

Now, let's run Kruskal's algorithm on the example graph:

```python
mst_edges_kruskal, min_weight_kruskal = kruskal_algorithm(graph)
print("Kruskal's Minimum Spanning Tree Edges: ", mst_edges_kruskal)
print("Kruskal's Minimum Weight: ", min_weight_kruskal)
```

```python
def kruskal_algorithm(graph):
    # 1. Create a list of all unique edges (weight, u, v)
    edges = []
    seen_edges = set() # To avoid adding (u,v) and (v,u) as separate edges in an undirected graph

    for u in graph:
        for v, weight in graph[u].items():
            # Create a canonical representation for undirected edges
            edge_key = tuple(sorted((u, v)))
            if (edge_key, weight) not in seen_edges:
                edges.append((weight, u, v))
                seen_edges.add((edge_key, weight))

    # Sort all edges in non-decreasing order of their weights
    edges.sort()

    # 2. Initialize disjoint set (Union-Find) structure
    parent = {vertex: vertex for vertex in graph}
    rank = {vertex: 0 for vertex in graph}

    def find(i):
        if parent[i] == i:
            return i
        parent[i] = find(parent[i]) # Path compression
        return parent[i]

    def union(i, j):
        root_i = find(i)
        root_j = find(j)
        if root_i != root_j:
            # Union by rank
            if rank[root_i] < rank[root_j]:
                parent[root_i] = root_j
            elif rank[root_i] > rank[root_j]:
                parent[root_j] = root_i
            else:
                parent[root_j] = root_i
                rank[root_i] += 1
            return True # Union successful
        return False # Already in the same set

    min_tree_edges = []
    min_weight = 0
    num_vertices = len(graph)
    num_edges_in_mst = 0

    # 3. Iterate through sorted edges
    for weight, u, v in edges:
        # Stop when V-1 edges are added (for a connected graph with V vertices)
        if num_edges_in_mst == num_vertices - 1:
            break

        if find(u) != find(v): # If adding this edge does not form a cycle
            union(u, v)
            min_tree_edges.append((u, v, weight))
            min_weight += weight
            num_edges_in_mst += 1

    return min_tree_edges, min_weight
```
