```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  # Description: This function calculates the sum of all elements in a given array.
  # Time Complexity: O(n) because it iterates through each element of the array once, where n is the number of elements in the array.
  total = 0
  for element in arr:
    total += element
  return total
```

```python
#call the function
arr = [12, 3, 4, 15,45,78,450]
print(sum_arr(arr))
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  # Description: This function finds the maximum element in a given array.
  # Time Complexity: O(n) because it iterates through each element of the array once, where n is the number of elements in the array.
  max_element = arr[0]
  for element in arr:
    if element > max_element:
      max_element = element
      return max_element
```
